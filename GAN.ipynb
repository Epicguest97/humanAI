```python
import os
from datasets import load_dataset, Dataset, DatasetDict
from huggingface_hub import HfApi
from PIL import Image

# Hugging Face repo details
HF_USERNAME = "Epicguest97"  # Your Hugging Face username
NEW_DATASET_NAME = "Spanish_documents_resized"  # New dataset name
LOCAL_DATASET_PATH = "resized_dataset"  # Local storage path
IMAGE_FOLDER = os.path.join(LOCAL_DATASET_PATH, "images")  # Image storage path

# Ensure output folders exist
os.makedirs(IMAGE_FOLDER, exist_ok=True)

# Load dataset
dataset = load_dataset("Epicguest97/Spanish_documents", split="train")

# Resize settings
IMAGE_SIZE = 256
Image.MAX_IMAGE_PIXELS = None  # Override decompression bomb error

# Process images safely
resized_data = []
for i, example in enumerate(dataset):
    try:
        img = example["image"]

        # Ensure the image isn't too large
        if img.size[0] * img.size[1] > 178956970:
            print(f"⚠️ Resizing large image {i}: {img.size} → ({IMAGE_SIZE}, {IMAGE_SIZE})")
            img = img.resize((IMAGE_SIZE, IMAGE_SIZE), Image.LANCZOS)

        # Save resized image
        img_filename = f"{i}.jpg"
        img_path = os.path.join(IMAGE_FOLDER, img_filename)
        img.save(img_path)

        # Store new dataset entry
        resized_data.append({"image": img_path})

    except Exception as e:
        print(f"❌ Error processing image {i}: {e}")

# Convert to Hugging Face Dataset
resized_dataset = Dataset.from_dict({"image": [d["image"] for d in resized_data]})
dataset_dict = DatasetDict({"train": resized_dataset})

# Save locally
dataset_dict.save_to_disk(LOCAL_DATASET_PATH)
```

```python
import os
import shutil

# Define the paths
parent_folder = "resized_dataset/images"
train_folder = os.path.join(parent_folder, "train")

# Create the "train" folder if it doesn't exist
os.makedirs(train_folder, exist_ok=True)

# Move all files from "images" to "images/train"
for file_name in os.listdir(parent_folder):
    file_path = os.path.join(parent_folder, file_name)

    # Check if it's a file (not a folder)
    if os.path.isfile(file_path):
        shutil.move(file_path, os.path.join(train_folder, file_name))

print("✅ All files moved to 'train' folder successfully!")
```
