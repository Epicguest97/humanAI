```python
import numpy as np
```

```python
from paddleocr import PaddleOCR
import cv2
import matplotlib.pyplot as plt

# Initialize PaddleOCR with DBNet (English model)
ocr = PaddleOCR(use_angle_cls=True, lang="en")  # Supports multiple languages

# Load your image
image_path = "page_2.jpg"  # Change this to your image path
img = cv2.imread(image_path)

# Perform text detection
result = ocr.ocr(image_path, cls=True)

# Draw bounding boxes on detected text
for line in result:
    for word in line:
        points = word[0]  # Get bounding box points
        cv2.polylines(img, [np.array(points, np.int32)], True, (0, 255, 0), 2)

# Show the result
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
